```python
import pandas as pd

# Load the full Guardians dataset
df_guardians = pd.read_csv('../CSVs/Guardians_Full_Dataset.csv')

# Display first few rows to confirm data
df_guardians.head()
```

```python
# Define valid events (ball in play only)
valid_events = [
    'single', 'double', 'triple', 'home_run', 'sac_fly', 'sac_fly_double_play', 'sac_bunt',
    'field_out', 'force_out', 'grounded_into_double_play', 'fielders_choice', 
    'double_play', 'fielders_choice_out'
]

# Filter dataset for only these events
df_guardians = df_guardians[df_guardians['events'].isin(valid_events)]

# Display first few rows to verify filtering
df_guardians.head()


```

```python
# Define the columns we want to keep
selected_columns = [
    'events', 'release_speed', 'pfx_z', 'plate_x', 'plate_z', 'vx0', 'vy0', 'vz0', 
    'ax', 'ay', 'az', 'sz_top', 'sz_bot', 'release_spin_rate', 
    'release_extension', 'release_pos_y', 'pitch_type'
]

# Keep only the selected columns
df_guardians = df_guardians[selected_columns]

# Display first few rows to confirm selection
df_guardians.head()

```

```python
# Define successful outcomes (hits)
successful_outcomes = ['single', 'double', 'triple', 'home_run', 'sac_fly', 'sac_fly_double_play', 'sac_bunt']

# Create the target variable (1 = hit, 0 = out)
df_guardians['target'] = df_guardians['events'].apply(lambda x: 1 if x in successful_outcomes else 0)

# Drop the original 'events' column since we no longer need it
df_guardians = df_guardians.drop(columns=['events'])

# Display first few rows to verify
df_guardians.head()
```

```python
# Check for missing values in each column
df_guardians.isna().sum()
```

```python
from sklearn.impute import SimpleImputer

# Define numerical columns that need imputation
columns_to_impute = ['release_spin_rate', 'release_extension']

# Initialize the imputer (using median strategy)
imputer = SimpleImputer(strategy="median")

# Apply imputation
df_guardians[columns_to_impute] = imputer.fit_transform(df_guardians[columns_to_impute])

# Verify missing values are handled
df_guardians.isna().sum()
```

```python
# Perform one-hot encoding on 'pitch_type'
df_pitch_encoded = pd.get_dummies(df_guardians['pitch_type'], prefix='pitch_type')

# Concatenate the new encoded columns with the main dataframe
df_guardians = pd.concat([df_guardians, df_pitch_encoded], axis=1)

# Drop the original 'pitch_type' column since it's now encoded
df_guardians = df_guardians.drop(columns=['pitch_type'])

# Display first few rows to verify
df_guardians.head()
```

```python
from sklearn.preprocessing import StandardScaler

# Define numerical columns to scale (excluding 'target')
numerical_columns = [
    'release_speed', 'pfx_z', 'plate_x', 'plate_z', 'vx0', 'vy0', 'vz0',
    'ax', 'ay', 'az', 'sz_top', 'sz_bot', 'release_spin_rate', 
    'release_extension', 'release_pos_y'
]

# Initialize the scaler
scaler = StandardScaler()

# Fit and transform numerical features
df_guardians[numerical_columns] = scaler.fit_transform(df_guardians[numerical_columns])

# Display first few rows to verify scaling
df_guardians.head()
```

```python
# Save the cleaned dataset to a CSV file
df_guardians.to_csv('Guardians_Cleaned_ML_All_ABs.csv', index=False)
```
